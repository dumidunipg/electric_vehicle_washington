# Clean for Visualizations

```python
EV_df = pd.read_csv('EV_Population_Clean.csv')
EV_df.head()
```

```python
# Get rid of missing data (0 values)
EV_df_clean = EV_df.copy() # making a copy for cleaning
EV_df_clean = EV_df_clean[EV_df_clean['Electric Range'].notna() & (EV_df_clean['Electric Range']>0)]
```

```python
#clean data for county visualization
county_count = EV_df_clean.groupby('County')['Electric Vehicle Type'].count().reset_index(name = 'vehicle_count')
filter_by_county = county_count[county_count['vehicle_count'] >= 1000]
EV_df_county_clean = EV_df[EV_df['County'].isin(filter_by_county['County'])]
EV_df_county_clean = EV_df_county_clean.drop(['VIN (1-10)', 'Postal Code', 'Model Year', 'Base MSRP', 'DOL Vehicle ID', 'Vehicle Location', '2020 Census Tract' ], axis = 1)
```

```python
# save file to csv to save space
EV_df_county_clean.to_csv('EV_df_county_clean.csv', index=False)
```
